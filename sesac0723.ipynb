### continue vs break
- continue: 다음 반복
- break: 반복문 탈출

```python
for i in range(5):
  if i==3:
    continue # 건너뛰기
  print(i)
```

```python
for i in range(5):
  if i ==3:
    break #반복 종료
  print(i)
```

```python
for j in range(3):
  for i in range(5):
    if i ==3:
      break # i만 멈춤 j는 안 멈춤
    print(j,i)
```
